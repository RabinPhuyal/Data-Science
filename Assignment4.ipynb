```python
hotel_data['adults'].fillna(hotel_data['adults'].mean(),inplace=True)
```

```python
hotel_data['adults'].isna().sum()
```

```python
for column in hotel_data.columns:
    if hotel_data[column].dtype=='object':
         hotel_data[column].fillna(hotel_data[column].mode()[0],inplace=True)
    else:
        hotel_data[column].fillna(hotel_data[column].mean(),inplace=True)
```

```python
hotel_data.isna().sum()
```

```python
duplicates=hotel_data.duplicated()
duplicates
```

```python
numeric_data=hotel_data.select_dtypes(include='number')

# Handle outliers using IQR method
for column in numeric_data.columns:
    Q1=numeric_data[column].quantile(0.25)
    Q3=numeric_data[column].quantile(0.75)
    IQR=Q3-Q1
    numeric_data=numeric_data[~((numeric_data[column] < (Q1 - 1.5 * IQR)) | (numeric_data[column] > (Q3 + 1.5 * IQR)))]

numeric_data

```

### Feature Engineering

```python
# Assuming booking_data is your DataFrame containing the data

# Calculate the duration of stay by adding 'stays_in_weekend_nights' and 'stays_in_week_nights'
hotel_data['stay_duration'] = hotel_data['stays_in_weekend_nights'] + hotel_data['stays_in_week_nights']

# Display the first few rows to verify the changes
print(hotel_data[['stays_in_weekend_nights', 'stays_in_week_nights', 'stay_duration']].head())
```

```python

# Derive new features from date columns
hotel_data['arrival_month'] = hotel_data['arrival_date'].dt.month_name()
hotel_data['arrival_day'] = hotel_data['arrival_date'].dt.day_name()
hotel_data['season'] = hotel_data['arrival_date'].dt.month % 12 // 3 + 1
hotel_data['season'] = hotel_data['season'].map({1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Fall'})

hotel_data.head()

```

### Handling Datetime Columns:

```python
# Convert reservation_status_date to datetime format with dayfirst=True
hotel_data['reservation_status_date'] = pd.to_datetime(hotel_data['reservation_status_date'], dayfirst=True)

# Extract additional features from the datetime column
hotel_data['reservation_status_day'] = hotel_data['reservation_status_date'].dt.day_name()
hotel_data['reservation_status_month'] = hotel_data['reservation_status_date'].dt.month_name()
hotel_data['reservation_status_hour'] = hotel_data['reservation_status_date'].dt.hour

# Display the first few rows to verify the changes
hotel_data[['reservation_status_date', 'reservation_status_day', 'reservation_status_month', 'reservation_status_hour']].head(10)
```
